## 🛠️ Tool Definitions


```python
# TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
```

```python
# TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        # Filter out short words and convert to lowercase unique keywords
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5] # Return up to 5 keywords
    except Exception:
        return []
```

## 🤖 Agent Logic Implementation


```python
# AGENT FUNCTION IMPLEMENTATION
import json

def agent(query: str):
    query_lower = query.lower()

    if "calculate" in query_lower:
        try:
            # Extract the expression after "calculate"
            expression_marker = "calculate"
            expression_start_index = query_lower.find(expression_marker) + len(expression_marker)
            expression = query[expression_start_index:].strip()

            if not expression:
                result = "No mathematical expression provided for calculation."
                response_type = "error"
            else:
                result = calculator(expression)
                response_type = "calculation"
        except Exception as e:
            result = f"Error in processing calculation: {e}"
            response_type = "error"
    elif "keywords from" in query_lower:
        try:
            # Extract the text after "keywords from"
            keywords_marker = "keywords from"
            keywords_text_start_index = query_lower.find(keywords_marker) + len(keywords_marker)
            text_for_keywords = query[keywords_text_start_index:].strip()

            if not text_for_keywords:
                result = "No text provided for keyword extraction."
                response_type = "error"
            else:
                result = extract_keywords(text_for_keywords)
                response_type = "keywords"
        except Exception as e:
            result = f"Error in processing keyword extraction: {e}"
            response_type = "error"
    else:
        result = f"Hello! I received your query: '{query}'. This is a general query and I can respond with information or assistance related to it."
        response_type = "general"

    return {
        "type": response_type,
        "result": result
    }
```

## 🧪 Test Cases


```python
queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?",
    "Calculate 10 * (5 + 2) / 3", # More complex calculation
    "Extract keywords from The quick brown fox jumps over the lazy dog", # Shorter text
    "Tell me about the weather forecast for tomorrow", # Another general query
    "calculate", # Test case for missing expression
    "extract keywords from" # Test case for missing text
]

for q in queries:
    print("Query:", q)
    response = agent(q)
    print("Response type:", response["type"])
    print("Response result:", response["result"])
    print("-" * 50)
```

## 🎯 Interactive Mode


```python
while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    response = agent(user_input)
    print("Response type:", response["type"])
    print("Response result:", response["result"])
    print("-" * 50)
```
